```python
import pandas as pd
```

```python
cols = ['index', 'url', 'state', 'code']
other_req = pd.read_csv("step3-othersites-requests.csv", header=None)
other_req.columns = cols
other_req["code"].fillna(-1, inplace=True)
other_req[["code"]] = other_req[["code"]].astype(int)

top_req = pd.read_csv("step3-topsites-requests.csv", header=None)
top_req.columns = cols
top_req["code"].fillna(-1, inplace=True)
top_req[["code"]] = top_req[["code"]].astype(int)
```

```python
othercounts = other_req.groupby(["state"], as_index=False).size().rename(columns={"size": "othersites"}).drop(columns=["state"])

topcounts = top_req.groupby(["state"], as_index=False).size().rename(columns={"size": "topsites"})

q10counts = pd.concat([topcounts, othercounts], axis=1)
q10counts["topsites"] /= q10counts["topsites"].sum()
q10counts["othersites"] /= q10counts["othersites"].sum()

q10counts.T.to_clipboard()

```

```python
import matplotlib.pyplot as plt
import numpy as np

other_periods = other_q4["period"]
top_periods = top_q4["period"]
min_period = min(other_periods.min(), top_periods.min())
max_period = max(other_periods.max(), top_periods.max())

print(f"min_period: {min_period}")
print(f"max_period: {max_period}")

bins = np.linspace(min_period, max_period, 20)

plt.hist(other_periods, bins, alpha=0.5, label='othersites')
plt.hist(top_periods, bins, alpha=0.5, label='topsites')
plt.xlabel("Certificate Validity Period (s)")
plt.ylabel("Number of Certificates")
plt.legend(loc='upper right')
plt.show()
print(bins)


# other_q4["period"].hist(bins=10)
# top_q4["period"].hist(bins=10)
```
